```python
import requests
from bs4 import BeautifulSoup

# Make a request to get a list of all cities in United States
page = requests.get("https://www.britannica.com/topic/list-of-cities-and-towns-in-the-United-States-2023068")
soup = BeautifulSoup(page.content, 'html.parser')

# ------------------------------------
# Get array with all states
# ------------------------------------

# 1D array that contains all states
states = []

# Set states to all <h2 class="h1"> tags of the soup
for element in soup.select('h2.h1'):
    states.append(element.text)

# ------------------------------------
# Get array with all cities split up by state
# ------------------------------------

# 1D array with all cities for current state
city_list = []

# 2D array with all cities split up by state
cities_list = [[]]

# counter to iterate through 'states' array
current_state = 0

# Set 'city' to all a.mdcrosslink tags of the soup
for city in soup.select('a.md-crosslink'):
    
    # if value at current index is equal to a state on the "states" list
    if city.text == states[current_state]:
        # add current list to full list
        cities_list.append(city_list)
        
        # reset city list
        city_list = []
        
        # increase state index counter 
        if current_state != len(states) - 1:
            current_state += 1

        # When 'city' is equal to the first state ('Alabama'),
        # clear the 'cities_list' array to remove anything that
        # isn't state/city
        if city.text == 'Alabama':
            cities_list = []
    else:
        # add current city to state_list
        city_list.append(city.text);
        
        if city.text == 'Worland':
            cities_list.append(city_list)
        
# Print all states + cities
for i in range(0,len(states)):
    print(states[i], cities_list[i], "\n") 
```
